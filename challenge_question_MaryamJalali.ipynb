## **Sara is a young woman who has just come to Berlin from Paris and wants to rent a suitable apartment in Berlin. We want to help her find an apartment.**

## The dataset:
The datasets contains information about rental appartments in Berlin. 
One row is for one appartment. The dataset contains 6 columns.  
- Region: is the Berlin district, where the flat is located
- Condition: is the condition of the flat
- Rooms: The number of rooms the flat has
- Rent: Monthly rent for the flat
- Year_Construction: The year in which the house was build
- Space: How many square meters does the flat has?

--- 


```
# Imports
import pandas as pd
```

```
# Read in dataset
rentals = pd.read_csv("https://raw.githubusercontent.com/maryam98/techstart_challenge/main/berlin_rental.csv")
rentals.head()
```

# . Where is the biggest and the smallest apartment?

```
#the biggest apartment :

rentals.nlargest(1,'Space')
```

```
#the smallest apartment:

rentals.nsmallest(1,'Space')
```

# . What is the mean rent?

```
rentals['Rent'].mean()
```

# .How many apartments is in **Mitte** in a **'well-kept'** condition?

```
rentals[(rentals.Region=="Mitte") & (rentals.Condition=="well_kept")]
```

```

rentals[(rentals.Region=="Mitte") & (rentals.Condition=="well_kept")]['Region'].count()
```

#.How many apartments are there in **Kreuzberg** built **before 1900**?

```
rentals[(rentals.Region=="Kreuzberg")&(rentals.Year_Construction<1900)]
```

```
rentals[(rentals.Region=="Kreuzberg")&(rentals.Year_Construction<1900)]['Region'].count()
```

#.Which is the most expensive **region** on average and what is the maximum **rent** for this region?

```
print('****average rent for each region:****')
print(rentals.groupby('Region')[['Rent']].mean())
print('\n****the most expensive region on average****')
a=rentals.groupby('Region')[['Rent']].mean()
ab=a.loc[a.Rent==a.Rent.max(),['Rent']]
print(ab)
print('\n**** what is the maximum rent for region',ab.index[0],'  ****')
rentals[rentals.Region==ab.index[0]]['Rent'].max()
```
